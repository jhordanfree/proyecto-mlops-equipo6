### Importar librerias

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score

import pandas as pd


import mlflow
import mlflow.sklearn
```

### Cargar el dataset

```python
iris = load_iris()

X = iris.data
y = iris.target

print(X.shape)
print(y.shape)
```

```python
df = pd.DataFrame(X, columns=iris.feature_names)

df["species"] = y

df
```

### Dividir el dataset

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    random_state=6
)

print(X_train.shape)
print(X_test.shape)
```

### Entrenamiento del modelo

```python
mlflow.set_experiment("Practica_MLOps_Iris_MLP")
```

```python

with mlflow.start_run():
    
    # Entrenamiento 
    modelo = MLPClassifier(random_state=6)
    modelo.fit(X_train, y_train)
    
    # Predicción y Evaluación del modelo
    predicciones = modelo.predict(X_test)
    accuracy = accuracy_score(y_test, predicciones)
    print(f"Accuracy del modelo: {accuracy:.4f}")
    
    # Inyección de MLOps 
    # Registramos los parámetros
    mlflow.log_param("hidden_layer_sizes", modelo.hidden_layer_sizes)
    mlflow.log_param("activation", modelo.activation)
    mlflow.log_param("solver", modelo.solver)
    mlflow.log_param("max_iter", modelo.max_iter)
    
    # Registramos la métrica final
    mlflow.log_metric("accuracy", accuracy)
    
    # Guardamos el artefacto
    mlflow.sklearn.log_model(
        modelo, 
        "modelo_mlp_clasificacion",
        skops_trusted_types=["sklearn.neural_network._stochastic_optimizers.AdamOptimizer"]
    )    
    print("¡Ejecución y tracking registrados en MLflow!")
```
